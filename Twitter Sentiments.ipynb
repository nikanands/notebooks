```python
sp_df = df.copy()
sp_df.head()
```

```python
for item in range(len(sp_df)):
    sp_df['text']= sp_df['text'].str.lower()
    sp_df['text'][item] = sp_df['text'][item].replace(str(sp_df['keyword'][item]),'')
```

```python
X = sp_df['text']
Y = sp_df['target']
```

```python
def create_nlp_doc(doc):
    return nlp(doc)
```

```python
docs = list(map(create_nlp_doc, sp_df.text))
```

```python
nlp.Defaults.stop_words |= {'<','>','br','/', '/>','</', '(', ')', '#'}
nlp.vocab['<'].is_stop = True
nlp.vocab['>'].is_stop = True
nlp.vocab['br'].is_stop = True
nlp.vocab['/'].is_stop = True
nlp.vocab['</'].is_stop = True
nlp.vocab['/>'].is_stop = True
nlp.vocab['('].is_stop = True
nlp.vocab[')'].is_stop = True
nlp.vocab['#'].is_stop = True
```

```python
def clean_review(doc):
    return[token.lemma_ for token in doc if token.is_stop == False and token.pos_ != 'PUNCT' and token.pos_ != 'NUM']
```

```python
sp_df['cleaned_review'] = list(map(clean_review, docs))
sp_df.head()
```

```python
import string
def filtered_doc(words):
    if words != string.punctuation:
        return ' '.join(words)
```

```python
sp_df['clean_doc'] = list(map(filtered_doc, sp_df.cleaned_review))
sp_df.head()
```

```python
X = sp_df['clean_doc']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=1)
```

```python
clf.fit(X_train, Y_train)
```

```python
predict = clf.predict(X_test)
```

```python
from sklearn.metrics import classification_report, accuracy_score
```

```python
print(classification_report(Y_test, predict))
```

```python
def remove_pattern(input_text, pattern):
    words = re.findall(pattern, input_text)
    for word in words:
        input_text = re.sub(word, '', input_text)
    return input_text.lower()
```

```python
nl_df['clean_tweet'] = np.vectorize(remove_pattern)(nl_df['text'], "@[\w]*")
nl_df.head()
```

```python
nl_df['clean_tweet'] = nl_df['clean_tweet'].str.replace("[^a-zA-Z#]", " ")
nl_df.head()
```

```python
nl_df['clean_tweet'] = nl_df['clean_tweet'].apply(lambda x: " ".join([w for w in x.split() if len(w) > 3]))
nl_df.head()
```

```python
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import train_test_split

from sklearn.metrics import accuracy_score, classification_report
from sklearn.naive_bayes import MultinomialNB
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
```

```python
nltk.download('stopwords')
stop_words = set(stopwords.words('english'))
```

```python
# Preprocessing the data
tokenized_tweet = []
def preprocess_text(tweets):
    tweet = []
    te = []
    punct_data = []
    clear_data = []
    
    for text in range(len(tweets)):
    # Converting into lowercase
        tweet.append(tweets[text].lower())
#         print(tweet)
    
    #Remove URLS
        te.append(re.sub(r"http\S+|www\S+|https\S+", "", tweets[text], flags=re.MULTILINE))
#         tweet.clear()
        
    # Remove Punctuations
        punct_data.append(te[text].translate(str.maketrans("", "", string.punctuation)))
        
    # Remove @ and #
        clear_data.append(re.sub(r'\@\w+|\#', "", punct_data[text]))
        
    # Tokenized tweet
        tokenized_tweet.append(word_tokenize(clear_data[text].lower()))
    return tokenized_tweet


preprocess_text(nl_df['text'])
```

```python
filtered_tweet = []
def clean_data(data):
    
    for words in data:
        a = []
        for word in words:
            
            if word not in stop_words:
                a.append(word)
                b = " ".join(a)
        filtered_tweet.append(b)
        
    return filtered_tweet

clean_data(tokenized_tweet)
```
