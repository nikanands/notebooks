```python
log_charges = np.log(data2['charges'])
```

```python
f, (ax1, ax2) = plt.subplots(1,2, sharey= True, figsize=(9,3))
plt.ticklabel_format(style='plain', axis='y')
ax1.scatter(data1['log_age'], log_charges)
ax1.set_title('Charges and Log Age')
ax2.scatter(data1['bmi'], log_charges)
ax2.set_title('Charges and Log BMI')
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
variables= data1[['smoker','region','children','age']]
vif = pd.DataFrame()
vif['VIF'] = [variance_inflation_factor(variables.values, i) for i in range(variables.shape[1])]
vif['Features'] = variables.columns
vif
```

```python
targets = log_charges
inputs = data2.drop(['charges', 'log_age', 'log_bmi', 'bmi','sex'], axis=1)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaler.fit(inputs)
```

```python
scaled_inputs = scaler.transform(inputs)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(scaled_inputs, targets, test_size=0.2, random_state=365)
```

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression()
reg.fit(x_train, y_train)
```

```python
y_hat = reg.predict(x_train)
```

```python
plt.scatter(y_train,y_hat)
plt.xlim(6,12)
plt.ylim(6,12)
plt.show()
```

```python
sns.distplot(y_train-y_hat)
plt.title('Residual PDF', size=20)
```

```python
reg.score(x_train, y_train)
```

```python
reg_summary = pd.DataFrame(inputs.columns.values, columns=['Features'])
reg_summary['Weights(coef_)'] = reg.coef_
reg_summary
```

```python
y_hat_test = reg.predict(x_test)
```

```python
plt.scatter(y_test,y_hat_test, alpha=0.2)
plt.xlim(6,12)
plt.ylim(6,12)
plt.show()
```

```python
df_pf = pd.DataFrame(np.exp(y_hat_test), columns=['Prediction'])
df_pf.head()
```

```python
y_test = y_test.reset_index(drop=True)
y_test
```

```python
df_pf['Target'] = np.exp(y_test)
df_pf
```

```python
df_pf['Residual'] = df_pf['Target'] - df_pf['Prediction']
```

```python
df_pf['Difference%'] = np.absolute(df_pf['Residual']/df_pf['Target']*100)
```

```python
pd.options.display.max_rows= 10000
df_pf.sort_values(by=['Difference%'])
```
