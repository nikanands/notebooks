```python
from statsmodels.stats.outliers_influence import variance_inflation_factor

variables= data2[['X2 house age','X3 distance to the nearest MRT station', 'X4 number of convenience stores']]

vif = pd.DataFrame()
vif['VIF'] = [variance_inflation_factor(variables.values, i) for i in range(variables.shape[1])]
vif['Features'] = variables.columns
vif
```

## Linear Regression

### Declaring targets and inputs with normal price then log price

```python
targets = log_price
inputs = data2.drop(['Y house price of unit area'], axis=1)
# inputs = [log_house, log_metro, log_store, log_lat, log_long]
```

## Scale the Data

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaler.fit(inputs)
```

```python
scaled_inputs = scaler.transform(inputs)
```

### Splitting the data for train and testing

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(scaled_inputs, targets, test_size=0.2, random_state=29)
```

### Creating the regression

```python
from sklearn.linear_model import LinearRegression

reg = LinearRegression()
reg.fit(x_train, y_train)
```

```python
y_hat = reg.predict(x_train)
```

```python
plt.scatter(y_train, y_hat)
plt.show()
```

```python
sns.displot(y_train-y_hat)
plt.title('Residual', size=20)
```

```python
reg.score(x_train, y_train)
```

```python
y_hat_test = reg.predict(x_test)
```

```python
plt.scatter(y_test, y_hat_test)
plt.show()
```

```python
df_pf = pd.DataFrame(np.exp(y_hat_test), columns=['Prediction'])
df_pf.head()
```

```python
y_test = y_test.reset_index(drop=True)
```

```python
df_pf['Target'] = np.exp(y_test)
df_pf
```

```python
df_pf['Residual'] = df_pf['Target'] - df_pf['Prediction']
```

```python
df_pf['Difference%'] = np.absolute(df_pf['Residual']/df_pf['Target']*100)
```

```python
pd.options.display.max_rows= 10000
df_pf.sort_values(by=['Difference%'])
```
